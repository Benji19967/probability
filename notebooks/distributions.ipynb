# Probability theory

## Generating data

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
# Example: generate data for weight of people
mean = 70
std_dev = 0.1
num_samples = 100
rng = np.random.default_rng()
data = rng.normal(mean, std_dev, num_samples)

sns.histplot(data, bins=30, kde=True, color="teal")
```

```python
# Joint pdf of weight and height -- assuming they are independent
rng = np.random.default_rng()

# weight
mean = 70
std_dev = 10
num_samples = 1000
weight = rng.normal(mean, std_dev, num_samples)

# height
mean = 175
std_dev = 15
num_samples = 1000
height = rng.normal(mean, std_dev, num_samples)

sns.histplot(x=weight, y=height, bins=30, kde=True, color="teal", cbar=True)
df = pd.DataFrame({"weight": weight, "height": height})
sns.pairplot(df)
```

```python
# Joint pdf of weight and height -- with correlation
rng = np.random.default_rng()

# 1. Define means for [weight, height]
means = [70, 175]

# 2. Define the Covariance Matrix
# [Var(weight), Cov(weight, height)]
# [Cov(height, weight), Var(height)]
cov_matrix = [
    [10**2, 120],  # 120 is the covariance (positive = correlated)
    [120, 15**2]
]

# 3. Generate the correlated data
data = rng.multivariate_normal(means, cov_matrix, size=1000)
weight = data[:, 0]
height = data[:, 1]

sns.histplot(x=weight, y=height, bins=30, kde=True, color="teal", cbar=True)
df = pd.DataFrame({"weight": weight, "height": height})
sns.pairplot(df)
```
