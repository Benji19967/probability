# Probability theory

## Generating data

```python
import numpy as np
import seaborn as sns
```

```python
# Example: generate data for weight of people
mean = 70
std_dev = 0.1
num_samples = 100
rng = np.random.default_rng()
data = rng.normal(mean, std_dev, num_samples)

sns.histplot(data, bins=30, kde=True, color="teal")
```
