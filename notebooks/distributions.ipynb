# Probability theory

## Generating data

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

rng = np.random.default_rng(seed=42)
```

```python
# Example generate data for weight of people
mean = 70
std_dev = 15
num_samples = 100
data = rng.normal(mean, std_dev, num_samples)

sns.histplot(data, bins=30, kde=True, color="teal")
plt.show()
```
